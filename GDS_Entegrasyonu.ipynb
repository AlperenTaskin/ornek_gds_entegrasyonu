## GDS'lerdeki Verilerin Tatilbudur Otelleri Arasindaki Karsiliklarini Buluyoruz 

```python
from fuzzywuzzy import fuzz
import numpy as np
```

Karsilastirmalar yapilirken 'fuzz.token_set_ratio' kullanilacak

```python
def skorla(otel_ad, otel_lokasyon, otel_url):
    skor = 0
    skor_listesi = []
    #bu katsayilar ahp calculator'dan geldi
    ad_katsayi = 0.659
    lokasyon_katsayi = 0.079
    url_katsayi = 0.263
    for index, row in tatilbudur_oteller.iterrows():
        ad_skoru = fuzz.token_set_ratio(row['OtelAdi'], otel_ad) * ad_katsayi
        lokasyon_skoru = fuzz.token_set_ratio(row['Lokasyon'], otel_lokasyon) * ad_katsayi
        url_skoru = fuzz.token_set_ratio(row['Url'], otel_url) * ad_katsayi
        skor = ad_skoru + lokasyon_skoru + url_skoru
        skor_listesi.append(skor)
        skor = 0
    #print(skor_listesi)
    index = np.argmax(skor_listesi)
    return tatilbudur_oteller.iloc[index]
```

```python
sonuc = skorla('Club Titan Hotel', 'Kargıcak, Antalya', 'Club-Titan-Hotel')
sonuc
```

## GDS1 Verilerinin Karsilastirilmasi

```python
for counter in range(len(gds1)):
    print('Gds1 verisindeki - '+ gds1.iloc[counter]['OtelAdi']+ ' - isimli otelin tatilbudur otelleri arasindaki karsiligi: - '+
          skorla(gds1.iloc[counter]['OtelAdi'], gds1.iloc[counter]['Lokasyon'], gds1.iloc[counter]['Url'])['OtelAdi']+' -\n')
```

## GDS2 Verilerinin Karsilastirilmasi

```python
for counter in range(len(gds2)):
    print('Gds2 verisindeki - '+ gds2.iloc[counter]['OtelAdi']+ '- isimli otelin tatilbudur otelleri arasindaki karsiligi: - '+
          skorla(gds2.iloc[counter]['OtelAdi'], gds2.iloc[counter]['Lokasyon'], gds2.iloc[counter]['Url'])['OtelAdi']+' -\n')
```
